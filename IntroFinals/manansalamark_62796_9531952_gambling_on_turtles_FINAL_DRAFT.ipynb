#turtle race


this program will line eight, different-colored turtles on the bottom of the screen, all set to a randomly selected speed. then they will race until one of those turtles reaches a certain height, to which the program will end, and thank the user. 

. . . the start and the end are for impression only

```python
#gambling on turtles
!pip3 install ColabTurtlePlus
import ColabTurtlePlus.Turtle as turtle
import time
import random

turtle.clearscreen()
turtle.initializeTurtle()
turtle.bgcolor("white")
turtle.hideturtle()
colors = ["red", "orange", "yellow", "green", "blue", "purple", "pink", "black"]

class racer(): #classes
  def __init__(self, turtle_color):
    self.racer_turtle = turtle.Turtle()
    self.racer_turtle.shape("Turtle")
    self.racer_name = str(turtle_color) + " Turtle"
    self.racer_color = self.racer_turtle.color(str(turtle_color))
    # self.racer_speed = self.racer_turtle.speed(random.randrange(13))
  def forward(self): #functions inside classes
    self.racer_turtle.forward(random.randrange(20))
  def left(self, deg):
    self.racer_turtle.left(deg)
  def jumpto(self, posX, posY):
    self.racer_turtle.jumpto(posX, posY)
  def getx(self):
    return self.racer_turtle.getx()
  def gety(self):
    return self.racer_turtle.gety()

red_turtle = racer("red")
orange_turtle = racer("orange")
yellow_turtle = racer("yellow")
green_turtle = racer("green")
blue_turtle = racer("blue")
purple_turtle = racer("purple")
pink_turtle = racer("pink")
black_turtle = racer("black")
all_turtles = [red_turtle, orange_turtle, yellow_turtle, green_turtle, blue_turtle, purple_turtle, pink_turtle, black_turtle]
all_colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'black']

def on_your_marks(): #functions outside of classes
  moving_constant = 0
  for t in all_turtles: #iterators
    t.jumpto(-350 + moving_constant, -100)
    t.left(90)
    moving_constant = moving_constant + 70

race_winner = 'None'

def start_race():
  while True: #conditional statements
    for racer in all_turtles: #for loop
      racer.forward() #geometric art

      x, y = racer.getx(), racer.gety()
      # print(x,y)
      if y >= 270:
        winner = all_colors[all_turtles.index(racer)]
        race_winner = winner
        return winner

#story intro
input("crack a joke: ") #printing, inputs, variables
print("you cringe a moment later. you say to yourself that it is time for bed")
time.sleep(2)
print("your vision flickers, and you enter a mindless space")
time.sleep(3)
print("you wake up, eyes crumbling, and of course the first thing you do when you wake up is...")
time.sleep(5)
print("to check your phone")
time.sleep(2)
print("elon has dm'ed you: you wanna gamble?")
time.sleep(3)

#elon and al gore
print("after a few conversations, elon directs you toward his friend, al gore")
time.sleep(4)
print("so... you want to make some money huh?")
time.sleep(3)
name = input("what should i call you?: ")
print("hello " + name + ".")
time.sleep(2)
print("to gamble, you need to connect your PayPal")
time.sleep(3)

#connecting paypal
print("Welcome to PayPal!")
time.sleep(1)
input("What is your username?: ")
input("And password?: ")
print("Processing...")
time.sleep(3)
print("For botting reasons, please verify that you are a human")
print("Please enter the following CAPTCHA: W82KL")
captcha = input("")
captcha_blockade = True
while captcha_blockade:
  if captcha == "W82KL":
    captcha_blockade = False
  else:
    print("Please try again.")
    print("Please enter the following CAPTCHA: W82KL")
    captcha = input("")

#welcoming
print("welcome to gambling!")
time.sleep(2)
print("in a moment, eight, different-colored turtles on the bottom of the screen")
time.sleep(2)
print("all set to a randomly selected speed.") 
time.sleep(5)
print("then they will race until one of those turtles reaches a certain height, or the finish line")
time.sleep(4)
print("your job is to pick which turtle will win.") 
time.sleep(3)
print("get it right and elon will gift you three doge coin")
time.sleep(3)
print("however, if you lose, you will lose your money AND elon will unfollow you...")
time.sleep(4)

#picking and starting
print("oh look above, the eight racers are lining up now!")
time.sleep(2)
on_your_marks()
time.sleep(3)
chosen_winner = input("Choose the winning turtle!: ")

choosing_blockade = True
while choosing_blockade: 
  if chosen_winner.lower() in colors:
    choosing_blockade = False
  else:
    print("that's not a racer, try again")
    chosen_winner = input("Choose the winning turtle!: ")
print("alright grab the popcorn...")
time.sleep(2)
print("on your marks...")
time.sleep(1)
print("get set")
time.sleep(1)
print("GOGOGO LOOK")
start_race()

#aftermath
if chosen_winner == race_winner:
  print("wow, you actually won. congradulations")
  time.sleep(2)
  print("you keep your elon follow, and you return home with stacks on stacks")
else: 
  print("yeah expected outcome. you lost")
  time.sleep(2)
  print("elon unfollows you, and you have to sell the house for money")
  time.sleep(2)
  print("moral of the story. dont gamble")

#ending
print("")
print("-------------")
print("thank you so much for playing " + name + "!")
print("this masterpiece was made by mark manansala")
print("-------------")

```
