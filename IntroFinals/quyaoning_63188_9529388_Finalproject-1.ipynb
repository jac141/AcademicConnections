I made this program to visualize the "adventure game" so that it can be more interesting. I also managed to put the words on the windows instead of printing it.

```python
# I created and ran this program with pycharm and copied it to the google colab after I finished, 
# and it seems not working with google colab

import turtle
import time


def pause():
    time.sleep(7)
#function

game = turtle.Screen()
game.title("maze")
game.bgcolor("black")
game.screensize(canvwidth=1920, canvheight=1080)
txt = turtle.Turtle()
#geometic art
txt.color("white")
txt.hideturtle()
style = ('Courier', 20, 'italic')
txt.write(" ")
time.sleep(3)
txt.write("You are trapped in an underground cave, navigate through it and get out", font=style, align='center')
pause()
txt.clear()
txt.write("You met a wizard", font=style, align='center')
pause()
txt.clear()
txt.color("blue")
txt.penup()
txt.goto(0, -300)
txt.write("Wizard: There's a present for you contained in one of the orbs", font=style, align='center')


class orb(turtle.Turtle):
    def __init__(self):
        turtle.Turtle.__init__(self)
        self.penup()
        self.shape("circle")
        self.color("yellow")
#class

orb1 = orb()
orb1.goto(0, 100)
orb2 = orb()
orb2.goto(-50, 0)
orb3 = orb()
orb3.goto(50, 0)
pause()
orb1.color("red")
txt.clear()
txt.write("Wizard: Only the red one contains the present", font=style, align='center')
time.sleep(3)
txt.clear()
txt.write("Wizard: Choose the correct one to get the present after I swap them around", font=style, align='center')
pause()
orb1.color("yellow")
orb2.goto(-100, 0)
orb3.goto(100, 0)
orb1.goto(0, 0)

time.sleep(2)
orb1.goto(-50, 50)
orb2.goto(-50, -50)
time.sleep(0.3)
orb1.goto(-100, 0)
orb2.goto(0, 0)

time.sleep(0.6)
orb2.goto(50, 50)
orb3.goto(50, -50)
time.sleep(0.3)
orb2.goto(100, 0)
orb3.goto(0, 0)

time.sleep(0.6)
orb1.goto(125, 75)
orb3.goto(-125, 75)
orb2.goto(0, 75)

time.sleep(1)
txt.clear()
txt.write("Wizard: Which one would you like, the left one(l), the middle one(m), or the right one(r)",
          font=('Courier', 15, 'italic'),
          align='center')
ans1 = input()
#variable

def fail():
    txt.color("white")
    txt.penup()
    txt.goto(0, 0)
    txt.clear()
    txt.write("You failed to escape the cave", font=style, align='center')
    time.sleep(4)
    exit()
#function

if ans1 != 'r':
    fail()

else:
    txt.color("white")
    txt.penup()
    txt.goto(0, 0)
    txt.write("You get a bottle of potion", font=style, align='center')
    time.sleep(3)
    txt.clear()
    orb1.goto(1080, 1920)
    orb2.goto(1080, 1920)
    orb3.goto(1080, 1920)
#conditional statement

txt.write("You walk forward and a gate blocks your way", font=style, align='center')
time.sleep(4)
txt.clear()
txt.write("You drink the potion and see the mechanical structure inside the gate", font=style, align='center')
time.sleep(4)
txt.clear()
turtle.register_shape("challenge2.gif")
gear = turtle.Turtle()
gear.shape("challenge2.gif")
time.sleep(2)
gear.hideturtle()
txt.write("You need to turn the largest red gear 6 rounds right by controlling the blue one with a handle attached to "
          "it to open the gate",
          font=('Courier', 11, 'bold'), align='center')
time.sleep(8)
txt.clear()
txt.write("Known the radius of the gears are 1 unit, 2 units, and 3 units respectively", font=style, align='center')
time.sleep(5)
txt.clear()
txt.penup()
txt.goto(0, -320)
txt.write("How should you operate the blue gear to open the gate? Indicate left and right with negative and positive, "
          "and rounds with numbers", font=('Courier', 11, 'bold'), align='center')
gear.showturtle()
ans2 = input()
game.bgcolor("black")
gear.hideturtle()
if ans2 != "+9":
    fail()
else:
    txt.up()
    txt.color("white")
    txt.goto(0, 0)
    txt.clear()
    txt.write("Congratulations, you escaped from the cave after walking through the gate", font=style, align='center')
    time.sleep(3)
    exit()
#conditional statement

turtle.done()
```
