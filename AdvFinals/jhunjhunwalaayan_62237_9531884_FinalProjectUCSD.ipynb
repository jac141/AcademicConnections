```python
!mkdir -p data#We are creating the data directories to reference the data from
#This is something I am new to and tried to implement in this program
!wget -nc https://nyc3.digitaloceanspaces.com/ml-files-distro/v1/investigating-sentiment-analysis/data/sentiment140-subset.csv.zip -P data
!unzip -n -d data data/sentiment140-subset.csv.zip
import pandas as pd
df = pd.read_csv("data/sentiment140-subset.csv", nrows=30000)
df.head()
```

```python
df.polarity.value_counts()
#polarity shows the number of 1s and number of 0s
#A 1 is classified as a positive tweet and a 0 is classified as a negative one
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(df.text)
words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
words_df.head()#Vectorizing helps break down data from matrices to a single vector. This is essential for sentiment analysis as we want data metrics per unique word
```

```python
X = words_df#Labelling our data
y = df.polarity
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

#I am going to compare two models to see which one gives the better accuracy and which one does not work as well
```

```python
logreg = LogisticRegression(C=1e9, solver='lbfgs', max_iter=1000)
logreg.fit(X, y)#Algo for logistic regression
```

```python
svc = LinearSVC()
svc.fit(X, y)#Algo for linear regression
```

```python
pd.set_option("display.max_colwidth", 200)#creating test data
unknown = pd.DataFrame({'content': [
    "I love love love love this kitten",
    "I hate hate hate hate this keyboard",
    "I'm not sure how I feel about toast",
    "Did you see the baseball game yesterday?",
    "The package was delivered late and the contents were broken",
    "Trashy television shows are some of my favorites",
    "I'm seeing a Kubrick film tomorrow, I hear not so great things about it.",
    "I find chirping birds irritating, but I know I'm not the only one",
]})
unknown
```

I had some trouble inititally with the implementation of the vectorizer feature as I wasn't sure of the syntax but after alot of crosschecking it started to work.

```python
print(vectorizer.get_feature_names())
unknown_vectors = vectorizer.transform(unknown.content)
unknown_words_df = pd.DataFrame(unknown_vectors.toarray(), columns=vectorizer.get_feature_names())
unknown_words_df.head()#Sets unknow_vectors with unknown content to prepare to test our models
```

```python
unknown['pred_logreg'] = logreg.predict(unknown_words_df)
unknown['pred_logreg_proba'] = logreg.predict_proba(unknown_words_df)[:,1]
unknown['pred_svc'] = svc.predict(unknown_words_df)
```

```python
#Testing models using fit
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X, y)
print("Training logistic regression")
logreg.fit(X_train, y_train)
print("Training SVC")
svc.fit(X_train, y_train)
#Logistic regression test
y_true = y_test
y_pred = logreg.predict(X_test)
matrix = confusion_matrix(y_true, y_pred)
label_names = pd.Series(['negative', 'positive'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names).div(matrix.sum(axis=1), axis=0)


```

I had to redo the confusion matrix part of this program as I was getting numbers and not percentages and had to crosscheck to get my code to output the testing metrics in percentages.

```python
#predicting linear regression
y_true = y_test
y_pred = svc.predict(X_test)
matrix = confusion_matrix(y_true, y_pred)
label_names = pd.Series(['negative', 'positive'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names).div(matrix.sum(axis=1), axis=0)
```

From this we can see that the SVC linear model is a better and we will go forward with that algorithim.
