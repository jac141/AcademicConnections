```python
def splitAndCount():
  string = input("Please enter your string... ")
  if len(string) > 10000:
    print("Your string can not be longer than 10,000 characters. Please try again.")
    splitAndCount()
  word_to_count = {}
  words = string.split()
  for i in words:
    words[words.index(i)] = "".join(e for e in i if e.isalnum())
  
  for i in words:
    word_to_count[i] = words.count(i)
  
  return word_to_count

print(splitAndCount())
```
